## Dictionaries
#### They are unordered, it uses the "key-value" to store data

```python
student_info = {"student_name":"John Doe", "age":20, "course":"python"}

print(student_info)
```

```python
type(student_info)
```

```python
#Accessing elements in a dictionary
#We use keys to access the elements in a dictionary

student_info["student_name"]
```

```python
student_info["age"]
```

```python
#You can store it in a variable
name = student_info["student_name"]
```

```python
age = student_info["age"]
```

```python
print(f"I am {name} and I am {age} years old. I studied {course} in the University")
```

```python
student_info["course"]
```

```python
course = student_info["course"]
```

```python
student_info.keys()
```

```python
student_info.values()
```

```python
student_info.items()
```

```python
#your value can be a list

student_info2 = {"student_name":"John Doe", "age":20, "course":["python", "SQL", "Geology"]}

print(student_info2)
```

```python
student_info2["course"]
```

```python
course_list = student_info2["course"]
```

```python
print(course_list)
```

```python
print(f" The last course is {course_list[2]}")
```

## conditional statements



To write a code to test whether a person is eligible to vote or not

```python
age = 20
```

```python
#using only if

if age >= 17:
    print("Yes, you are eligible to vote")
    print("congratulations")
    
```

```python
age2 = 14
```

```python
#using if else

if age2 >= 17:
    print("Yes, you are eligible to vote")
    print("congratulations")
else:
    print("oops, you are not eligible to vote")
    
```

**To test whether a number is zero, positive or negative**

```python
num = 12
```

```python
#using if-elif: elif is used when you want to test for multiple conditions in a variable

if num == 0:
    print("The number is zero")
elif num < 0:
    print("The number is negative")
elif num > 0:
    print("The number is positive")
    
```

To test whether a number is even or odd

```python
test_num = 2
```

```python
#Remember the modulus sign% returns the remainder after division

if (test_num % 2) == 0:
    print("The number is even")
elif (test_num % 2) == 1:
    print("The number is odd")
```

```python
test_num2 = 47
```

```python
if test_num2 % 2 == 0:
    print("The number is even")
elif test_num2 % 2 == 1:
    print("The number is odd")
```

```python
greeting = "Hello everyone"
```

```python
#using in

#To test whether the word Hi is in a greeting

if "Hi" in greeting:
    print("The user used Hi")
else:
    print("The user did not use Hi")
```

```python
#list of a womans cars

cars = ["Audi", "Toyota", "Benz", "Tesla"]
```

```python
if "Tesla" in cars:
    print("She has a Tesla")
else:
    print("She does not have Tesla")
```

```python
if "Tesla" in cars:
    print("She has a Tesla")
elif "Benz" in cars:
    print("She has a Benz")
else:
    print("Maybe she has other models")
```

```python
#if the conditions you want to pass are all true, use multiple "if's" rather than using elif, Python will then run all the "if's"

if "Tesla" in cars:
    print("She has a Tesla")
if "Benz" in cars:
    print("She has a Benz")
else:
    print("Maybe she has other models")
```

## Nested If's

```python
print(cars)
```

```python
len(cars)
```

```python
if len(cars) > 0: #this is checking if she at least has a car
    if "Benz" in cars: #checking if Benz is among the cars she has, if at all she has a car
        print("She has a Benz")
    else: #this one is saying if the above conditions are contrary then, print the sentence below
        print("She does not have a Benz")
else:  #this one is saying if she doesn't meet the very first condition print the sentence below
    print("She doesn't have a car")
```
